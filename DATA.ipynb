```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('DATA.xlsx',parse_dates=True,index_col='Date')
```

```python
df.head()
```

```python
df.info()
```

```python
len(df)
```

```python
test_perc=14
```

```python
test_index=len(df)-test_perc
```

```python
train=df.iloc[:test_index]
```

```python
test=df.iloc[test_index:]
```

```python
train
```

```python
test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(train)
```

```python
scaled_train=scaler.transform(train)
```

```python
scaled_test=scaler.transform(test)
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length=3
generator=TimeseriesGenerator(scaled_train,scaled_train,length,batch_size=1)
```

```python
X,y=generator[0]
```

```python
X
```

```python
y
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
n_features=11
model=Sequential()
model.add(LSTM(100,activation='relu',input_shape=(length,n_features)))
model.add(Dense(scaled_train.shape[1]))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
val_generator=TimeseriesGenerator(scaled_test,scaled_test,length=length,batch_size=1)
```

```python
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(monitor='val_loss',patience=2)
```

```python
model.fit_generator(generator,epochs=10,validation_data=(val_generator),callbacks=[early_stopping])
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
import numpy as np
```

```python
test_predictions=[]
first_eval_batch=scaled_train[-length:]
current_batch=first_eval_batch.reshape((1,length,n_features))

for i in range(len(test)):
    current_pred=model.predict(current_batch)[0]
    
    test_predictions.append(current_pred)
    current_batch=np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
true_predictions=scaler.inverse_transform(test_predictions)
```

```python
test.shape
```

```python
true_predictions.shape
```

```python
true_predictions_df=pd.DataFrame(np.round(true_predictions),index=test.index,columns=['MP','TP','CP','OP','YP','BP','KP','EP','FP','SP','RP',])
```

```python
new=pd.concat([test,true_predictions_df],axis=1)
```

```python
new['Milk']==new['MP']
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(scaled_test,test_predictions))
```

```python
full_scaler=MinMaxScaler()
```

```python
full_scaled=full_scaler.fit_transform(df)
```

```python
generator=TimeseriesGenerator(full_scaled,full_scaled,length=length,batch_size=1)
```

```python
n_features=11
model=Sequential()
model.add(LSTM(100,activation='relu',input_shape=(length,n_features)))
model.add(Dense(scaled_train.shape[1]))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit_generator(generator,epochs=3)
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
forecast=[]
periods=2
first_eval_batch=scaled_train[-length:]
current_batch=first_eval_batch.reshape((1,length,n_features))

for i in range(periods):
    current_pred=model.predict(current_batch)[0]
    
    forecast.append(current_pred)
    current_batch=np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
forecast=scaler.inverse_transform(forecast)
```
